# Slicing examples
There are *many* ways to do this. 

To find out more: 
http://pandas.pydata.org/pandas-docs/stable/indexing.html

```python
titanic[10:15]
```

### Selecting rows
~~~
data[start:end]
~~~
- Missing values imply "from the start" or "to the  end"
- Negative values imply "from the end"

### Selecting columns by name 
~~~
data['column']
data[['column1', 'column2']]
~~~

```python
titanic[['age', 'class', 'fare']].head()
```

### Selecting both rows and columns using iloc

The basic format is 
~~~
data.iloc[start_row:end_row, start_column:end_column]

data.iloc[[list of rows by index], [list of columns by index]]
~~~

# Selecting based on search criteria (Boolean Indexing)

~~~
data[mask]
~~~

Where mask is a true-false array we can get by a boolean expression. 
~~~
mask = dataset['column'] > 10
~~~

Keep only observations witha age larger than 50

```python
titanic[ (titanic['age'] > 50) & (titanic['sex'] == 'male')].head()
```

# Changing the table

## Setting explicit value
~~~
titanic.iloc[row_id,column_ids] = value (or [values...])
titanic.loc[row,column_names] = value (or [values...])
~~~

## Appending a column

~~~
data.loc[row_range, column] = data 
~~~
Data can be a single value or a whole column

# Groupby
Splits the dataframe into groups by the specified column(s)
Each group has its own dataframe with its members. 

https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.groupby.html

More info on this operations:
http://pandas.pydata.org/pandas-docs/stable/groupby.html

We can then apply aggregation functions per group
~~~
grouped_data.mean()
~~~

```python
# gender/class stats
titanic.groupby(['class', 'who']).mean()
```
